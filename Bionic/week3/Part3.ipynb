## 5. 代码示例 -2 机器人路径规划

### 5.1 问题描述

在本节中，我们将使用遗传算法来训练主管控制一组机器人搜索环境，以覆盖尽可能多的区域。

在这种情况下，我们考虑问题的三个要素：
- 我们可以命令一组机器人在环境中导航。 
- 我们希望探索的环境，包含不同位置的框。 
- 一种用于指挥进化的机器人的控制策略。

**环境：**

我们将检查简单的室内环境，如下所示。环境被蓝色方块覆盖，表示机器人需要达到的目标。

**机器人：**

在这里，我们考虑一组可以被命令向不同方向移动的三个机器人。这些机器人可以共同被命令分散并朝八个不同的方向前进。机器人将朝这些方向前进，同时相互排斥和障碍物。此外，我们可以控制机器人之间的传播，使它们可以非常紧密地包装或分散得很远。

**控制策略：**

给定一个要探索的环境和一组要控制的机器人，任务是制定一种控制策略，命令机器人实现最佳覆盖。在这里，我们将使用行为树来表示我们进化的基因组。一棵树将包含我们将通过进化优化的有序命令序列。我们通过查看机器人组中心的位置及其分布来使树做出决策。下面显示了一个树的示例。

![](./pics/robot.png)

**如何作用？**

对于每棵树，从左到右依次触发每个节点。当到达绿色的条件节点时，如果条件为 FALSE，则序列将停止并从头重新开始。这允许我们根据生成的条件触发不同阶段的命令。

![决策树](./pics/behave_tree.png)

上面的树会首先命令机器人向北行驶，而 y 方向的质心不大于 25，机器人将继续向北行驶。 一旦声明属实，机器人就会被引导向南。

总而言之，进化的目标是找到最佳的动作序列和条件来控制机器人组达到所有目标。
